```python
import requests 
import time
from bs4 import BeautifulSoup
import re
from requests_html import HTMLSession 

url = requests.get('https://caroleasylife.blogspot.com/search/label/..%E8%A5%BF%E9%BB%9E%3A%3A%E8%9B%8B%E7%B3%95%28%E6%88%9A%E9%A2%A8%E6%B5%B7%E7%B6%BF%29?updated-max=2020-04-24T15:44:00%2B08:00&max-results=1000&start=20&by-date=false')
time.sleep(0.5)
soup = BeautifulSoup(url.text)
alla = soup.select("h3 > a" )

# # cake = re.compile(r'戚風蛋糕')
# # right = cake.findall('alla')

# session = HTMLSession()
# r = session.get('https://caroleasylife.blogspot.com/search/label/..%E8%A5%BF%E9%BB%9E%3A%3A%E8%9B%8B%E7%B3%95%28%E6%88%9A%E9%A2%A8%E6%B5%B7%E7%B6%BF%29?updated-max=2020-04-24T15:44:00%2B08:00&max-results=1000&start=20&by-date=false')
# article_entries = r.html.find('a')

for i in alla:
    url = i.attrs['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    title = soup.find('h3').text
    sub_title = re.sub('\(.*?\)', '', title)
    article = soup.find('div',{'class': 'post-body'} ).text
    m = re.findall('材料:.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*\n.*', article)
    turn = "".join(m)
    c = turn.split()
    turnstr = "".join(c)
    sub_1 = re.sub('步驟.*', '', turnstr)
    sub_2 = re.sub('..準備工作.*', '', sub_1)
    sub_3 = re.sub('\(.*?\)', '', sub_2)
    
    print(sub_title , sub_3)
    
```
